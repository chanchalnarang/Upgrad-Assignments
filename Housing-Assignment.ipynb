```python
## importing all libraries

import numpy as np
import pandas as pd
import sklearn as sl
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as se
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading and Understanding and visualising the Data

housing_Data=pd.read_csv(r'Housing.csv')
housing_df=pd.DataFrame(housing_Data)
print(housing_df.head())
print ('standard info about dataset')
housing_df.info()
print('provides stastical information about dataset')
housing_df.describe()
print('provide no of rows and columns')
housing_df.shape
y=housing_df['price']
x=housing_df['furnishingstatus']
plt.title('Relationship between price Vs furnishingstatus' ,fontdict={'color':'black'})
plt.xlabel('\n furnishingstatus',fontdict={'color':'black'}) 
plt.ylabel('price',fontdict={'color':'black'})
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.bar(x,y)
plt.show()







    
```
